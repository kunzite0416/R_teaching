```r
# 安裝並載入 ggplot2 套件
# install.packages("ggplot2")
library(ggplot2)

# 建立簡單的資料框
df <- data.frame(x = -40:40, y = (-40:40)^2)

# 繪製散佈圖
ggplot(df, aes(x = x, y = y)) +
  geom_point() +
  ggtitle("Simple scatter plot")
```

# **資料處理與視覺化實作**

目標： 使用 babynames 套件分析美國嬰兒名字的流行趨勢。​

步驟：

*   安裝並載入 babynames 套件。​
*   篩選特定名字的資料。​
*   繪製名字流行趨勢圖。​

```r
# 安裝並載入 babynames 套件
#install.packages("babynames")
library(babynames)

# 篩選名字為 "Bruno" 的資料
bruno_df <- subset(babynames, name == "Bruno")

# 繪製名字流行趨勢圖
plot(bruno_df$year, bruno_df$n, type = "l",
     main = "Bruno 名字的流行趨勢",
     xlab = "年份", ylab = "人數")
```
